## Step 4: Applying the Analytics Lifecycle

**Problem Statement:** Does money buy happiness? (GDP vs Happiness analysis)


````python
# DESCRIPTIVE ANALYTICS: Happiness scores by region
regional_stats = df.groupby('region').agg(
    avg_happiness=('happiness_score', 'mean'),
    avg_gdp=('gdp_per_capita', 'mean'),
    country_count=('country', 'count')
).round(2).sort_values('avg_happiness', ascending=False)

print('=== HAPPINESS BY REGION (Descriptive Analytics) ===')
display(regional_stats)
````

````python
# DIAGNOSTIC ANALYTICS: Which factors correlate most with happiness?
numeric_cols = ['gdp_per_capita', 'social_support', 'life_expectancy',
                'freedom_score', 'corruption_perception']

correlations = (
    df[numeric_cols + ['happiness_score']]
    .corr()['happiness_score']
    .drop('happiness_score')
)

print('=== FACTORS CORRELATED WITH HAPPINESS (Diagnostic Analytics) ===')
for factor, corr in correlations.items():
    direction = 'Positive' if corr > 0 else 'Negative'
    print('  {:30s}: {:+.3f}  ({})'.format(factor, corr, direction))
````

````python
# VISUALIZATION: GDP vs Happiness scatter plot
plt.figure(figsize=(12, 7))

regions = df['region'].unique()
colors = plt.cm.Set2(range(len(regions)))
color_map = dict(zip(regions, colors))

for region in regions:
    mask = df['region'] == region
    plt.scatter(
        df[mask]['gdp_per_capita'] / 1000,
        df[mask]['happiness_score'],
        label=region, s=80, alpha=0.8,
        color=color_map[region]
    )

for _, row in df.iterrows():
    if row['happiness_score'] > 7.3 or row['happiness_score'] < 4.0 or row['country'] == 'India':
        plt.annotate(
            row['country'],
            (row['gdp_per_capita'] / 1000, row['happiness_score']),
            textcoords='offset points', xytext=(5, 5), fontsize=8
        )

plt.xlabel('GDP per Capita (in $000s)', fontsize=12)
plt.ylabel('Happiness Score (0-10)', fontsize=12)
plt.title('Does Money Buy Happiness?\nGDP per Capita vs. Happiness Score', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print('Insight: Positive correlation, but diminishing returns above ~$50K!')
````

## Step 5: Lab Exercises

Complete each exercise. The solution is already provided  study it, then try writing it yourself in a new cell.


````python
# EXERCISE 1: Find the top 5 happiest countries and their GDP
top5_happy = df.nlargest(5, 'happiness_score')[['country', 'happiness_score', 'gdp_per_capita']]
display(top5_happy)
````

````python
# EXERCISE 2: Which African countries are in the top 50% happiness globally?
median_happiness = df['happiness_score'].median()
happy_african = df[
    (df['region'] == 'Africa') & (df['happiness_score'] >= median_happiness)
]
print('African countries above global median happiness (', round(median_happiness, 2), '):')
print(happy_african[['country', 'happiness_score']].to_string(index=False))
````

````python
# EXERCISE 3: Bar chart of average happiness by region
region_avg = df.groupby('region')['happiness_score'].mean().sort_values(ascending=True)

plt.figure(figsize=(10, 5))
plt.barh(region_avg.index, region_avg.values, color='steelblue', alpha=0.8)
plt.axvline(df['happiness_score'].mean(), color='red', linestyle='--', label='Global Avg')
plt.xlabel('Average Happiness Score')
plt.title('Average Happiness Score by Region')
plt.legend()
plt.tight_layout()
plt.show()
````
